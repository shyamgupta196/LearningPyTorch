The SMAF Takes in input from a NN layer and converts them into probabilites .

As You have seen from the image below the entry with the highest value will have the highest probability.

Thats it this is what softmax does

It takes in input of An Array and converts them into probabilities all you have to do is to take <b>np.argmax</b>(if an array) of it

```python
# Lets Get to the Numpy Version CODE! of it
import numpy as np

def softmax(array):
    return np.exp(array)/np.sum(np.exp(array))
```

```python
array = np.array([2,1.0,0.1],dtype=np.float32)
```

```python
output = softmax(array)
print(f'output: {output}')
```

##### well you see that just what it is , if you need the max of the output's too then just use np.argmax

```python
array[np.argmax(output)]
```

```python
# Torch Version of the Softmax Looks different 
import torch 
array = torch.tensor([2,1,0.1],dtype=torch.float32)
loss = torch.softmax(array,dim=0)
print(f'loss: {loss}')
```
